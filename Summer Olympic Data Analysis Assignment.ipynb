# **2.Which sport is having most number of Gold Medals so far? (Top 5)**

```python
sport = df["Sport"].unique()
```

```python
df1 =df[df["Medal"]=="Gold"]
df1.head()
```

```python
data2 = []
for i in sport:
    data2.append([i,len(df1[df1["Sport"]==i])])
```

```python
data2
```

```python
data2 = pd.DataFrame(data2,columns=["Sports","Gold Medal"])
data2 = data2.sort_values(by= "Gold Medal",ascending=False).head(5)
data2.plot(x = "Sports",y = "Gold Medal",kind = "bar",figsize =(13,6),xlabel="Sports",ylabel = "Frequency",rot=15)
for index, value in enumerate(data2["Gold Medal"]):
    plt.text(index,value,str(value))
```

**8. Which sport is having most female Gold Medalists? (Top 5)**

```python
df8 = df[df["Gender"]=="Women"]
df8 = df8[df8["Medal"]=="Gold"]
```
